```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_iris = pd.read_csv('Iris.csv')
data_iris
```

```python
data_iris.info()
```

```python
data_iris.describe()
```

```python
print(data_iris['Species'].unique())
```

```python
from sklearn.preprocessing import LabelEncoder
for col in ['Id','SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species']:
    encoder = LabelEncoder()
    data_iris[col] = encoder.fit_transform(data_iris[col])
    data_iris
    print(data_iris['Id'], data_iris['SepalLengthCm'], data_iris['SepalWidthCm'], data_iris['PetalLengthCm'], data_iris['PetalWidthCm'],data_iris['Species'])
```

```python
data_iris
```

```python
x = data_iris.drop('Species',axis=1)
y = data_iris['Species']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.3, random_state=42)
```

```python
print(x_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kkn = KNeighborsClassifier(n_neighbors=1)
```

```python
kkn.fit(x_train, y_train)
```

```python
print("Akurasi score sebesar: {}".format(kkn.score(x_test,y_test)*100))
```

```python
y_prediksi = kkn.predict(x_test)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_prediksi)
```

```python
print("Prediksi:",y_prediksi)
print("Akurasi: {:.2f}%".format(accuracy * 100))
```
